```python

import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=META&interval=1min&month=2022-01&outputsize=full&apikey=ZP3FYW2GOPLQPDGX'
r = requests.get(url)
data = r.json()

time_series = data.get("Time Series (1min)", {})
print(time_series)
```

```python
import pandas as pd

df = pd.read_json(data)
df.to_csv('202001_IBM.csv', encoding='utf-8', index=False)
```
